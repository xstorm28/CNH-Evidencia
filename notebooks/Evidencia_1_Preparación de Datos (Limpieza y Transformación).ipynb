#**2. Esquemas de datos**:

1. **Fuentes de Datos:**
- Telematrix: Se dispone de 4 conjuntos de datos (`Mantenimientos 2024-2025`, `Indicadores de Ruteo`, `Horas 2024-2025`, `Reporte_unidades_dia_anterior`) provenientes de este servicio de telemetría que ofrece como servicio la empresa CNH. Esta fuente proporciona datos detallados sobre el estado y operación de las unidades específicas del proyecto.
- Control Room: Se utiliza una base de datos (`PopulationView_2026`), la cual proviene de una plataforma de telemetría global. Esta fuente ofrece una visión global o de referencia que no guarda relación directa con las unidades individuales de las otras bases.


---

2. **¿Cómo se obtiene la información?**
- Origen y Transmisión: La información es capturada originalmente mediante sensores de telemetría en campo. Estos datos son procesados por los sistemas de CNH y Control Room.

- Protocolo de Acceso: En nuestro caso, el acceso a la información se realizó mediante archivos en formato xlsx brindados por el Socio Formador en una carpeta de Drive.

- Herramientas de Análisis: El procesamiento, limpieza y visualización de los datos actualmente se está realizando utilizando la librería Pandas en Python, lo que permite una manipulación eficiente.


---

3. **Esquema de Datos y Relacionales:**
- Identificador principal (Primary Key): Para las 4 bases de datos provenientes de Telematrix, el campo clave es el Identificador de Unidad de CNH (VIN). Este campo permite cruzar la información de telemetría entre los distintos archivos.

- Datos Aislados: La base de datos `PopulationView_2026` funciona como un conjunto de datos independiente. Debido a que es una base de datos global, se nos especifícó en los archivos compartidos que esta contiene equipos diferentes sin manera de conectarse con los de Telematrix, aunque igualmente están identificados por el VIN y el número de serie del dispositivo telemático instalado en la unidad.

Para esta exploración inicial, nos estaremos enfocando en 3 pilares que nos ayudaran a comprender de mejor manera como es el desempeño-historial que nos proporciona telemetria

### Perfilado de Uso y Desgaste (Rangos Operativos)

**Distribución de Carga:**

```python
import pandas as pd
import numpy as np

# Obtenemos los descriptores estadísticos clave
# Usamos el horómetro como métrica principal de desgaste
stats_horas = df_unidades['horometro'].describe()

# Extraemos los cuartiles para definir los límites de nuestras categorías
q1 = df_unidades['horometro'].quantile(0.25)
q3 = df_unidades['horometro'].quantile(0.75)
mediana = df_unidades['horometro'].median()

# Definimos una función para categorizar
def categorizar_uso(horas):
    if horas <= q1:
        return 'Uso Bajo (Subutilizada)'
    elif horas <= q3:
        return 'Uso Estándar (Rango Típico)'
    else:
        return 'Uso Intensivo (Desgaste Acelerado)'

# Aplicamos la categorización al dataframe
df_unidades['categoria_uso'] = df_unidades['horometro'].apply(categorizar_uso)

# Resumen estadístico para justificar el valor real
print("--- Límites de Categorización (Basados en Estadística) ---")
print(f"Límite Uso Bajo: Menos de {q1:.2f} hrs")
print(f"Rango Típico (50% de la flota): Entre {q1:.2f} y {q3:.2f} hrs")
print(f"Límite Uso Intensivo: Más de {q3:.2f} hrs")

print("\n--- Distribución de la Flota por Categoría ---")
print(df_unidades['categoria_uso'].value_counts(normalize=True) * 100)
```

En flotas de maquinaria, el promedio suele estar "inflado" por unidades viejas o completamente nuevas. Al usar el Rango Intercuartílico ($IQR = Q_3 - Q_1$), identificamos el comportamiento del 50% central de la flota, que es lo que realmente define la operación típica.

Ademas de esto, nos puede dar una mejor idea de donde estan paradas las viejas y las nuevas. Las unidades en Uso Bajo representan dinero desperdiciado (maquinaria parada). Mientras que las de uso Intensivo son las que necesitan inspección inmediata para evitar fallas catastróficas.

**Calidad Operativa (Ratio Km/Hr - Ralentí).**

```python
# Calculamos la relación de movimiento
# Usamos un pequeño valor para evitar división entre cero
df_unidades['ratio_movimiento'] = df_unidades['odometro'] / (df_unidades['horometro'] + 0.1)

# Descriptores estadísticos del ratio
stats_ratio = df_unidades['ratio_movimiento'].describe()

# Categorización de Eficiencia de Uso
def categorizar_ritmo(ratio):
    if ratio < 5:
        return 'Operación Estática / Ralentí Alto'
    elif ratio <= 25:
        return 'Operación Mixta (Normal)'
    else:
        return 'Operación de Traslado / Alta Movilidad'

df_unidades['perfil_ritmo'] = df_unidades['ratio_movimiento'].apply(categorizar_ritmo)

print("--- Distribución de Perfiles de Ritmo de Trabajo ---")
print(df_unidades['perfil_ritmo'].value_counts(normalize=True) * 100)

# Vamos a ver qué tan bajo es
print("--- Detalle Estadístico del Ratio de Movimiento ---")
print(df_unidades['ratio_movimiento'].describe())

# Veamos cuántas máquinas tienen un ratio cercano a CERO absoluto
maquinas_muertas = len(df_unidades[df_unidades['ratio_movimiento'] < 0.1])
print(f"\nMáquinas con movimiento casi nulo (<0.1): {maquinas_muertas}")
```

Aqui hay algo muy interesante, el 98% de la flota aparece como perfil en ralenti alto, pero no quiere decir obligatoriamente que no se estan moviendo. Un tractor arando se mueve a 5-8 km/h. Si el ratio es 2 o 3, entraría en tu categoría de "Estática", pero en realidad está produciendo.

**Intensidad Mensual**

```python
# Calcular antigüedad en meses (ajustado a la fecha actual del sistema)
from datetime import datetime
hoy = datetime.now()

# Aseguramos que fecha_alta sea datetime
df_unidades['fecha_alta'] = pd.to_datetime(df_unidades['fecha_alta'])

# Calculamos meses (usamos 30.44 como promedio de días al mes)
df_unidades['meses_antigüedad'] = ((hoy - df_unidades['fecha_alta']).dt.days / 30.44).round(0)
df_unidades['meses_antigüedad'] = df_unidades['meses_antigüedad'].replace(0, 1) # Evitar división entre cero

# Intensidad de Uso (Horas que suma el tractor cada mes de su vida)
df_unidades['intensidad_mensual'] = df_unidades['horometro'] / df_unidades['meses_antigüedad']

# Categorización de "Explotación del Activo"
q_int = df_unidades['intensidad_mensual'].quantile([0.25, 0.75])

def categorizar_intensidad(valor):
    if valor <= q_int[0.25]: return 'Uso Extensivo (Bajo)'
    if valor <= q_int[0.75]: return 'Uso Operativo (Normal)'
    return 'Sobreexplotación (Crítico)'

df_unidades['estatus_explotacion'] = df_unidades['intensidad_mensual'].apply(categorizar_intensidad)

print("--- Resumen de Intensidad Mensual (Horas/Mes) ---")
print(df_unidades['estatus_explotacion'].value_counts(normalize=True) * 100)
```

* Sobreexplotación (29.4%): Casi 1 de cada 3 tractores está trabajando a un ritmo frenético. Si un tractor promedio debería sumar, digamos, 100 horas al mes, estos están sumando 200 o más.
  * Aqui el valor puede esar en justamente predecir o medir cuando ocuparan mantenimiento estas maquinarias para poder monetizar y de igual manera estos deben ser uno de nuestros principales objetivos
  * Deben ser los primeros en recibir llamadas para contratos de mantenimiento prepagado. Si se están "acabando" el tractor, van a necesitar piezas (filtros, aceites, correas) mucho antes que el resto. Es ingreso asegurado a corto plazo.
* Uso Extensivo / Bajo (23.5%): Estas unidades son "dinero durmiendo". Tienen meses y casi no suman horas. Por lo tanto si no se usan, no ocupan mantenimiento constante

Debido a esto, se vuelve necesario priorizar procesos de limpieza de datos, como la normalización de valores, la eliminación de inconsistencias y, en algunos casos, la separación o creación de nuevas columnas que permitan mejorar la calidad de los enlaces. Estas acciones pueden ayudar a fortalecer las relaciones entre datasets y reducir posibles errores en etapas posteriores.

De igual manera, identificamos que columnas descriptivas como `marca` y `modelo` están presentes en la mayoría de los datasets, lo cual permite realizar agrupaciones, segmentaciones y análisis comparativos entre diferentes tipos de equipos. No obstante, estas variables no garantizan unicidad, ya que múltiples registros pueden compartir la misma marca o modelo. Por esta razón, su uso como clave de relación debe ser complementario y no principal, ya que podrían generar duplicidades o relaciones incorrectas si se utilizan de forma aislada.

Asimismo, existen otras columnas como `latitud`, `longitud`, `ciudad` y `estado`, que están presentes en datasets como **df_unidades** y **df_horas_2024**. Estas variables permiten realizar análisis geográficos y segmentaciones espaciales, lo cual puede ser útil para identificar patrones de comportamiento por región o ubicación. Sin embargo, no generan una relación directa entre registros, ya que no son identificadores únicos y pueden repetirse en múltiples casos. Su valor radica más en el análisis contextual que en la unión de datos e informació

Finalmente, columnas como **distribuidor** pueden generar relaciones a nivel organizacional, permitiendo vincular información entre datasets desde una perspectiva de negocio. No obstante, este tipo de relación suele ser de tipo uno a muchos (one-to-many) más que uno a uno (one-to-one), ya que un mismo distribuidor puede estar asociado a múltiples registros o unidades. Por lo tanto, aunque estas columnas son útiles para análisis agregados y segmentación, no deben considerarse como claves principales de enlace, sino como variables complementarias dentro del modelo de datos.

### **5. Detectar valores atípicos que podrían afectar el análisis para su posterior transformación (no en este avance). Ejemplo: Puedes utilizar el método del IQR (rango intercuartílico) para identificar outliers. Genera un boxplot o scatterplot bivariado para visualizar su impacto en la relación entre variables.**

```python
def outliers_plot(df, df_name):
    numeric_cols = df.select_dtypes(include=np.number).columns

    for col in numeric_cols:
        q1  = df[col].quantile(0.25)
        q3  = df[col].quantile(0.75)
        iqr = q3 - q1
        min_inlier  = q1 - 1.5 * iqr
        max_inlier  = q3 + 1.5 * iqr
        outliers = df[(df[col] < min_inlier) | (df[col] > max_inlier)][col].count()
        print(f"{df_name} columna a analizar : {col} :Cuántil 1: {q1:.2f}, Cuántil 3: {q3:.2f}, IQR: {iqr:.2f}, Outliers: {outliers}\n")

    n = len(numeric_cols)
    ncols = 3
    nrows = (n + ncols - 1) // ncols
    fig, axes = plt.subplots(nrows, ncols, figsize=(15, nrows * 4))
    axes = axes.flatten()

    for i, col in enumerate(numeric_cols):
        axes[i].boxplot(df[col].dropna())
        axes[i].set_title(col, fontsize=9)
        axes[i].set_xticks([])

    for j in range(i + 1, len(axes)):
        axes[j].set_visible(False)

    plt.suptitle(f'Boxplots — {df_name}', fontsize=13)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

La fucnión que se encuentra en la parte superior es una función que se creo debido a que necesitábamos para graficar las boxplots de todas las columnas de tipo númerico por dataframe. Creamos la función outliers_plot que va a contar el número de outliers y grafica. Lo primeoro que se hizo fue meter todas las columnas númeircas en un ciclo for para que las leyera, y con esto obtener de cada una las variables de : **q1,q3,iqr,min_inlier,max_inlier**.

Es importante destacar que la función recibe dos datos, cómo se llama el df y el string de cómo aparecera nombrado el dataframe al realizar el plt.suptitle.

El cálculo de los outliers se calcula de la siguiente manera: **outliers = df[(df[col] < min_inlier) | (df[col] > max_inlier)][col].count()**

Esto nos ayuda a detectarlos y saber qué tan sesgada puede llegar a estar nuestra base de datos.

* **expl_cerrados_promedio**: Esta columna junto con los pendientes nos permite calcular el ratio de riesgo. Si un segmento tiene muchos pendientes pero pocos cerrados, la red de distribuidores no está cubriendo la demanda de ese grupo.
* **expl_ratio_riesgo**: Esta columna es la métrica central del Riesgo de Incumplimiento. Divide los pendientes promedio entre los cerrados promedio de cada segmento. Un ratio alto significa que por cada servicio que se cierra hay varios que se quedan sin atender, lo cual indica alta probabilidad de pérdida de clientes.
* **expl_intensidad_promedio**: Esta variable contextualiza el riesgo con la intensidad de uso. Un segmento con alto ratio de riesgo y alta intensidad es el más crítico, porque son tractores que trabajan mucho, necesitan servicio con urgencia y no lo están recibiendo.

```python
df_total['expl_pendientes_promedio'] = df_total.groupby('estatus_explotacion')['pendientes'].transform('mean')

df_total['expl_cerrados_promedio'] = df_total.groupby('estatus_explotacion')['cerrados'].transform('mean')

df_total['expl_ratio_riesgo'] = df_total['expl_pendientes_promedio'] / (df_total['expl_cerrados_promedio'] + 1)

df_total['expl_intensidad_promedio'] = df_total.groupby('estatus_explotacion')['intensidad_mensual'].transform('mean')
```

## `Discretización o Binning` con el Ordinal Encoding

```python
mapeos = {
    'horometro_bin': {
        'Nuevo (0-500)': 1, 'En uso (500-2000)': 2,
        'Alto uso (2000-5000)': 3, 'Crítico (5000+)': 4
    },
    'antiguedad_bin': {
        'Reciente (0-2 años)': 1, 'Mediano (2-3 años)': 2,
        'Maduro (3-4 años)': 3, 'Antiguo (4+ años)': 4
    },
    'intensidad_bin': {
        'Baja (0-25)': 1, 'Media (25-60)': 2,
        'Alta (60-120)': 3, 'Muy alta (120+)': 4
    },
    'combustible_bin': {
        'Bajo (0-10)': 1, 'Medio (10-50)': 2,
        'Alto (50-100)': 3, 'Muy alto (100+)': 4
    }
}

# Aplicamos el mapeo SOBRE la misma columna para no crear nuevas
for col, mapping in mapeos.items():
    # Convertimos a string por seguridad antes del mapeo
    df_total[col] = df_total[col].astype(str).map(mapping).fillna(0).astype(int)
```

```python
df_total.info()
```

Justificación-Explicación:

**Resolución y Estructura:** La variable numérica le da al modelo la capacidad de encontrar cortes quirúrgicos (ej. separar tractores con exactamente $2450.5$ horas). El bin ordinal le da una "pista" estructural sobre grupos de riesgo ya conocidos por el negocio.

**Tratamiento de Redundancia**: A diferencia de una Regresión Lineal, a XGBoost no le afecta tener información redundante. Si el bin no aporta nada nuevo sobre el valor numérico, el algoritmo simplemente le dará una importancia de cero y usará la otra.

**No linealidad:** A veces, el riesgo no sube de forma constante con el número. Los bins ayudan al modelo a entender que hay saltos bruscos de "estatus" entre un rango y otro.

Por eso se crean los bins ya categorizados y aparte dejar la columna original (donde esta el dato numerico)

Por lo general, el Mantenimiento Preventivo es el estado "normal" o planificado, mientras que el Mantenimiento Correctivo representa la falla o la reparación no programada.


Si es una característica (Feature): Ayuda al modelo a entender el historial del tractor. Por ejemplo, un tractor con muchos correctivos previos tiene más riesgo de fallar pronto.

Si es el Target (y): El modelo aprenderá a predecir si el próximo servicio será preventivo o correctivo basándose en la telemetría. (prueba de hipotesis que tenemos)

```python
"""
DIAGNÓSTICO MSNO — Identificación de equipos con registros faltantes
Proyecto CNH · Equipo: Alina, Ian, Daniel, Eugenio

Corre este script en tu notebook de Google Colab después de cargar df_total.
Genera tablas con los alias, modelos y distribuidores que tienen NaN
en cada grupo de columnas del MSNO.
"""

import pandas as pd
import missingno as msno
import matplotlib.pyplot as plt

# ─────────────────────────────────────────────────────────────
# BLOQUE 1 — Grupos de columnas según el MSNO
# (basado en la imagen que nos compartiste)
# ─────────────────────────────────────────────────────────────

GRUPOS = {
    "Identidad del equipo": [
        "imei", "alias", "marca", "modelo", "configuracion",
        "fecha_alta", "num_serie", "distribuidor", "ciudad", "estado"
    ],
    "Telemetría principal": [
        "horometro", "odometro"
    ],
    "Servicios y mantenimiento": [
        "total", "cerrados", "c_fuera", "pendientes",
        "hora_50", "hora_300", "hora_600", "hora_900",
        "hora_1200", "hora_1500", "hora_1800", "hora_2100", "hora_2400"
    ],
    "Datos de contacto": [
        "cliente", "telefono", "email"
    ],
    "Telemetría operativa": [
        "latitud", "longitud", "combustible", "distancia",
        "vel_max_kph", "vel_promedio_kph",
        "t_encendido_hrs", "t_apagado_hrs"
    ],
    "Variables construidas": [
        "categoria_uso", "ratio_movimiento", "perfil_ritmo",
        "meses_antigüedad", "intensidad_mensual", "estatus_explotacion"
    ],
}

# Filtrar solo columnas que existen en el dataframe
GRUPOS_VALIDOS = {
    grupo: [c for c in cols if c in df_total.columns]
    for grupo, cols in GRUPOS.items()
}

# ─────────────────────────────────────────────────────────────
# BLOQUE 2 — Resumen global de NaN por columna
# ─────────────────────────────────────────────────────────────

print("=" * 65)
print("RESUMEN GLOBAL — % de registros faltantes por columna")
print("=" * 65)

nan_resumen = (
    df_total.isnull()
    .sum()
    .reset_index()
    .rename(columns={"index": "columna", 0: "nan_count"})
)
nan_resumen["pct_faltante"] = (nan_resumen["nan_count"] / len(df_total) * 100).round(1)
nan_resumen = nan_resumen[nan_resumen["nan_count"] > 0].sort_values("pct_faltante", ascending=False)

print(nan_resumen.to_string(index=False))
print(f"\nTotal de filas en df_total: {len(df_total):,}")

# ─────────────────────────────────────────────────────────────
# BLOQUE 3 — Equipos sin telemetría operativa
# (latitud, longitud, vel_max_kph, etc. → columna oscura en MSNO)
# ─────────────────────────────────────────────────────────────

cols_telemetria = GRUPOS_VALIDOS.get("Telemetría operativa", [])

if cols_telemetria:
    mask_sin_telemetria = df_total[cols_telemetria].isnull().any(axis=1)
    df_sin_telemetria = df_total[mask_sin_telemetria][
        ["alias", "modelo", "distribuidor", "estado"] + cols_telemetria
    ].drop_duplicates(subset=["alias"])

    print("\n" + "=" * 65)
    print(f"EQUIPOS SIN TELEMETRÍA OPERATIVA ({len(df_sin_telemetria)} únicos)")
    print("Columnas afectadas:", cols_telemetria)
    print("=" * 65)
    print(df_sin_telemetria[["alias", "modelo", "distribuidor", "estado"]].to_string(index=False))

# ─────────────────────────────────────────────────────────────
# BLOQUE 4 — Equipos sin datos de contacto
# (cliente, telefono, email → patrón de bloque vacío en MSNO)
# ─────────────────────────────────────────────────────────────

cols_contacto = GRUPOS_VALIDOS.get("Datos de contacto", [])

if cols_contacto:
    mask_sin_contacto = df_total[cols_contacto].isnull().all(axis=1)
    df_sin_contacto = df_total[mask_sin_contacto][
        ["alias", "modelo", "distribuidor", "estado"] + cols_contacto
    ].drop_duplicates(subset=["alias"])

    print("\n" + "=" * 65)
    print(f"EQUIPOS SIN NINGÚN DATO DE CONTACTO ({len(df_sin_contacto)} únicos)")
    print("Columnas afectadas:", cols_contacto)
    print("=" * 65)
    print(df_sin_contacto[["alias", "modelo", "distribuidor", "estado"]].to_string(index=False))

# ─────────────────────────────────────────────────────────────
# BLOQUE 5 — Equipos sin historial de mantenimiento por umbral
# (hora_300 a hora_2100 → zona más fragmentada del MSNO)
# ─────────────────────────────────────────────────────────────

cols_umbrales = [c for c in [
    "hora_300", "hora_600", "hora_900", "hora_1200",
    "hora_1500", "hora_1800", "hora_2100"
] if c in df_total.columns]

if cols_umbrales:
    # Contar cuántos umbrales tiene NULL cada equipo
    df_total["umbrales_faltantes"] = df_total[cols_umbrales].isnull().sum(axis=1)

    df_umbrales_criticos = (
        df_total[df_total["umbrales_faltantes"] >= 4]
        [["alias", "modelo", "distribuidor", "estado",
          "horometro", "umbrales_faltantes"] + cols_umbrales]
        .drop_duplicates(subset=["alias"])
        .sort_values("umbrales_faltantes", ascending=False)
    )

    print("\n" + "=" * 65)
    print(f"EQUIPOS CON 4+ UMBRALES DE MANTENIMIENTO FALTANTES ({len(df_umbrales_criticos)})")
    print("=" * 65)
    print(df_umbrales_criticos[
        ["alias", "modelo", "distribuidor", "estado", "horometro", "umbrales_faltantes"]
    ].to_string(index=False))

# ─────────────────────────────────────────────────────────────
# BLOQUE 6 — Perfil completo de NaN por equipo
# Equipos que acumulan NaN en MÚLTIPLES grupos simultáneamente
# ─────────────────────────────────────────────────────────────

print("\n" + "=" * 65)
print("PERFIL DE NaN POR EQUIPO — equipos con más columnas faltantes")
print("=" * 65)

df_total["total_nan_columnas"] = df_total.isnull().sum(axis=1)

df_perfil_nan = (
    df_total[df_total["total_nan_columnas"] > 5]
    [["alias", "modelo", "distribuidor", "estado",
      "horometro", "total_nan_columnas"]]
    .drop_duplicates(subset=["alias"])
    .sort_values("total_nan_columnas", ascending=False)
    .head(30)
)

print(df_perfil_nan.to_string(index=False))

# ─────────────────────────────────────────────────────────────
# BLOQUE 7 — NaN por distribuidor
# ¿Cuál distribuidor genera más registros incompletos?
# ─────────────────────────────────────────────────────────────

print("\n" + "=" * 65)
print("NaN POR DISTRIBUIDOR — quién genera más dark data estructural")
print("=" * 65)

nan_por_dist = (
    df_total.groupby("distribuidor")["total_nan_columnas"]
    .agg(["mean", "count", "sum"])
    .rename(columns={
        "mean": "nan_promedio_por_equipo",
        "count": "total_equipos",
        "sum": "nan_total_acumulado"
    })
    .round(1)
    .sort_values("nan_promedio_por_equipo", ascending=False)
    .head(15)
    .reset_index()
)

print(nan_por_dist.to_string(index=False))

# ─────────────────────────────────────────────────────────────
# BLOQUE 8 — Visualización MSNO con grupos marcados
# ─────────────────────────────────────────────────────────────

fig, ax = plt.subplots(figsize=(20, 10))
msno.matrix(
    df_total.sample(min(500, len(df_total)), random_state=42),
    ax=ax,
    sparkline=True,
    color=(0.08, 0.24, 0.39),
    fontsize=9
)
ax.set_title(
    "MSNO — Mapa de valores faltantes (muestra 500 equipos)\n"
    "Zonas blancas = NaN · Cada columna = una variable del df_total",
    fontsize=12, pad=15
)
plt.tight_layout()
plt.savefig("msno_diagnostico_completo.png", dpi=200, bbox_inches="tight")
plt.show()
print("\nImagen guardada: msno_diagnostico_completo.png")

# ─────────────────────────────────────────────────────────────
# BLOQUE 9 — Exportar a Excel para revisión manual
# ─────────────────────────────────────────────────────────────

with pd.ExcelWriter("diagnostico_nan_equipos.xlsx", engine="openpyxl") as writer:

    nan_resumen.to_excel(writer, sheet_name="Resumen NaN por columna", index=False)

    if cols_telemetria and not df_sin_telemetria.empty:
        df_sin_telemetria.to_excel(writer, sheet_name="Sin telemetría", index=False)

    if cols_contacto and not df_sin_contacto.empty:
        df_sin_contacto.to_excel(writer, sheet_name="Sin contacto", index=False)

    if cols_umbrales and not df_umbrales_criticos.empty:
        df_umbrales_criticos.to_excel(writer, sheet_name="Umbrales faltantes", index=False)

    df_perfil_nan.to_excel(writer, sheet_name="Perfil NaN por equipo", index=False)
    nan_por_dist.to_excel(writer, sheet_name="NaN por distribuidor", index=False)

print("\nArchivo exportado: diagnostico_nan_equipos.xlsx")
print("Hojas generadas:")
print("  · Resumen NaN por columna")
print("  · Sin telemetría")
print("  · Sin contacto")
print("  · Umbrales faltantes")
print("  · Perfil NaN por equipo")
print("  · NaN por distribuidor")
```

```python
# Columnas de umbrales de horas que aparecen en el MSNO
HORAS = [c for c in [
    "hora_50","hora_300","hora_600","hora_900",
    "hora_1200","hora_1500","hora_1800","hora_2100","hora_2400"
] if c in df_total.columns]

# Para cada IMEI: cuáles horas tiene en blanco
df_imei = df_total[["imei","alias","modelo","distribuidor","estado"] + HORAS].copy()

# Columna que dice cuántas horas están en blanco
df_imei["horas_en_blanco"] = df_imei[HORAS].isnull().sum(axis=1)

# Columna que lista exactamente cuáles están en blanco
df_imei["cuales_en_blanco"] = df_imei[HORAS].apply(
    lambda row: ", ".join([col for col in HORAS if pd.isnull(row[col])]),
    axis=1
)

# Solo los que tienen AL MENOS UNA hora en blanco
df_resultado = (
    df_imei[df_imei["horas_en_blanco"] > 0]
    .sort_values("horas_en_blanco", ascending=False)
    .reset_index(drop=True)
)

print(f"Total IMEI con horas en blanco: {len(df_resultado)}")
print(f"Total IMEI en el dataset:       {len(df_total)}\n")

# Tabla resumen: cuántos IMEI por número de horas faltantes
print("── Distribución ──────────────────────────")
print(df_resultado["horas_en_blanco"].value_counts().sort_index(ascending=False).to_string())
print()

# Lista completa
cols_ver = ["imei","alias","modelo","distribuidor","estado","horas_en_blanco","cuales_en_blanco"]
print("── Lista completa de IMEI ────────────────")
print(df_resultado[cols_ver].to_string(index=False))

# Exportar a Excel en Drive
ruta = "/content/drive/MyDrive/CNH2026/imei_horas_faltantes.xlsx"
df_resultado[cols_ver].to_excel(ruta, index=False)
print(f"\nExportado a: {ruta}")
```
